# Tiền xử lý dữ liệu

```python
# chỉnh lại kiểu dữ liệu cho một số biến nhị phân và biến phân loại
binary_columns = []
for column_name in data:
    unique_values = data[column_name].unique()
    if unique_values.shape[0] == 2:
        binary_columns.append(column_name)

numerical_columns = ['PhysHlth', 'MentHlth', 'BMI']

categorical_columns = set(data.columns) - set(numerical_columns) - set(binary_columns)

binary_categorical_columns = binary_columns + list(categorical_columns)

data[binary_categorical_columns] = data[binary_categorical_columns].astype(int)
```

```python
path = os.path.join('/kaggle/working', 'data_clean.csv')
data_clean = data.copy()

data_clean.to_csv(path)
```
